# Confronto tra algoritmi di ordinamento: Selection Sort vs Quick Sort

In questo notebook confrontiamo due algoritmi di ordinamento fondamentali:

- **Selection Sort**: algoritmo semplice, ma inefficiente per grandi dataset.
- **Quick Sort**: algoritmo più complesso, ma molto più veloce nella pratica.

Analizzeremo:
- Implementazione degli algoritmi
- Complessità computazionale
- Confronto delle prestazioni su array casuali
- Visualizzazione dei risultati
- Analisi del caso peggiore del quicksort
- possibile soluzione per evitare il caso peggiore del quicksort


## Import delle librerie necessarie

```python
# Import delle librerie necessarie
import time
import random
import matplotlib.pyplot as plt
```

## Implementazione degli algoritmi

```python
def selection_sort(A):
    n = len(A)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if A[j] < A[min_index]:
                min_index = j
        A[i], A[min_index] = A[min_index], A[i]
```

```python
def quicksort(A, p, r):
    if p < r:
        q = partition(A, p, r)
        quicksort(A, p, q - 1)
        quicksort(A, q + 1, r)

def partition(A, p, r):
    x = A[r]  # pivot
    i = p - 1
    for j in range(p, r):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[r] = A[r], A[i + 1]
    return i + 1
```

## Spiegazione degli Algoritmi

- **Selection Sort**: è un algoritmo di ordinamento semplice e intuitivo. Funziona trovando, ad ogni iterazione, il valore minimo nella parte non ordinata dell’array e scambiandolo con il primo elemento di quella parte. Questo processo si ripete spostando il limite della parte ordinata verso destra, fino a che tutto l’array è ordinato. Si nota così che la sua complessità sarà sempre O($n^2$)


- **Quick Sort**: è un algoritmo di ordinamento molto meno intuitivo e semplice. L’idea principale è quella di scegliere un elemento pivot, dividere l’array in due parti in base al pivot, e ricorsivamente ordinare le due sottosequenze. La sua complessità varia in base se si trovi nel caso migliore, medio e peggiore. Il suo vantaggio sta nel fatto che nel caso migliore e medio avrà una complesssità di O($nlogn$), purtroppo nel caso peggiore (quando cerca di ordinare dei dati già ordinati) ci sarà un peggioramento a O($n^2$)

## Funzione per misurare il tempo di esecuzione

```python
def time_algorithm(algorithm, dim):
    start = time.time()
    temp=dim.copy()
    if algorithm == selection_sort:
        algorithm(temp)
    else:
        algorithm(temp, 1, len(temp) - 1)
    end = time.time()
    return end - start
```

## Confronto delle prestazioni 

```python
sizes = [100, 200, 400, 800, 1600, 2000, 4000, 8000, 10000]
selection_times = []
quick_times = []

for size in sizes:
    data = [random.randint(0, 10000) for _ in range(size)]
    selection_times.append(time_algorithm(selection_sort, data))
    quick_times.append(time_algorithm(quicksort, data))
```

## Visualizzazione dei risultati

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, selection_times, label="Selection Sort", marker='o')
plt.plot(sizes, quick_times, label="Quick Sort", marker='o')
plt.title("Confronto tra Selection Sort e Quick Sort")
plt.xlabel("Dimensione dell'array")
plt.ylabel("Tempo di esecuzione (secondi)")
plt.legend()
plt.grid(True)
plt.show()
```

## Analisi della complessità

**Selection Sort**
- Complessità temporale: O(n²) in tutti i casi
- Complessità spaziale: O(1)

**Quick Sort**
- Complessità temporale media: O(n log n)
- Caso peggiore: O(n²) (quando l'array è già ordinato e si sceglie sempre il primo elemento come pivot)
- Complessità spaziale: O(log n) a causa della ricorsione


## Analisi caso peggiore del Quick Sort

Il Quick Sort, come spiegato precedentemente, nel caso peggiore raggiunge una complessità temporale di $O(n^2)$. Questo avviene quando, durante ogni chiamata ricorsiva, il pivot scelto non divide l’array in modo bilanciato, ma produce una partizione molto sbilanciata, come ad esempio:

- Un lato con $n - 1$ elementi
- L'altro con 1 elemento

Se l’array è già ordinato (crescente o decrescente) e si sceglie sempre come pivot il primo o l’ultimo elemento, la partizione sarà molto inefficiente, ogni chiamata ricorsiva riduce l’array solo di un elemento, generando una profondità di ricorsione pari a n, e quindi un numero totale di confronti proporzionale a: O($n^2$)
 

## Visulizzazione della situazione

```python
sizes= [100, 200, 400, 800, 1600, 2000]
worstQuickTimes=[]

for size in sizes:
    data = list(range(size))  # caso peggiore
    worstQuickTimes.append(time_algorithm(quicksort, data))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, worstQuickTimes, label="QuickSort - Caso Peggiore", marker='o', color='red')
plt.xlabel("Dimensione dell'array")
plt.ylabel("Tempo di esecuzione (secondi)")
plt.title("QuickSort - Caso Peggiore (Array Ordinato)")
plt.grid(True)
plt.legend()
plt.show()
```

come si può notare ha un andamento esponenziale

## Soluzione al caso peggiore

una possibile strategia per evitare di finire nel caso peggiore del Quick Sort è **randomizzare la scelta del pivot**

Ecco la sua implementazione: 

```python
def randomized_quicksort(A, p, r):
    if p < r:
        q = randomized_partition(A, p, r)
        randomized_quicksort(A, p, q - 1)
        randomized_quicksort(A, q + 1, r)

def randomized_partition(A, p, r):
    i = random.randint(p, r)   # Sceglie un pivot casuale
    A[r], A[i] = A[i], A[r]    # Scambia il pivot con l'ultimo elemento
    return partition(A, p, r)  # Partizione classica come nel QuickSort
```

## Valutiamo il Random Quick Sort

```python
sizes= [100, 200, 400, 800, 1600, 2000]
randomQuickTimes=[]
worstQuickTimes=[]
for size in sizes:
    data = list(range(size))
    randomQuickTimes.append(time_algorithm(randomized_quicksort, data))
    worstQuickTimes.append(time_algorithm(quicksort, data))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, randomQuickTimes, label="Worst Case QuickSort Random", marker='o')
plt.plot(sizes, worstQuickTimes, label="Worst Case Quick Sort", marker='o')
plt.title("Valutazione nel caso peggiore del Quick Sort Randomized")
plt.xlabel("Dimensione dell'array")
plt.ylabel("Tempo di esecuzione (secondi)")
plt.legend()
plt.grid(True)
plt.show()
```

Vediamo che l'algoritmo **Quick Sort Randomized** è molto più efficiente del **Quick Sort** normale nel caso peggiore (array già ordinato)

## Vediamo nel dettaglio l'andamento del Quick Sort Randomized

```python
sizes= [100, 800, 1600, 2000, 4000, 8000, 10000, 16000, 20000, 50000, 100000]
randomQuickTimes=[]
for size in sizes:
    data = list(range(size))
    randomQuickTimes.append(time_algorithm(randomized_quicksort, data))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, randomQuickTimes, label="Worst Case QuickSort Random", marker='o')
plt.title("Andamento del Quick Sort Randomized quando abbiamo un array ordinato")
plt.xlabel("Dimensione dell'array")
plt.ylabel("Tempo di esecuzione (secondi)")
plt.legend()
plt.grid(True)
plt.show()
```

notiamo che ha un andamento simil O($nlogn$)

## Conclusione

- Il Quick Sort è significativamente più veloce su dataset di grandi dimensioni.
- Selection Sort è più semplice da implementare, ma inefficiente per grandi volumi di dati.
- Quick Sort è la scelta preferita nella pratica, anche se bisogna gestire bene il caso peggiore, però un vantaggio del Quick Sort è che è facilmente estendibile rendendo gestibile anche il caso peggiore.
